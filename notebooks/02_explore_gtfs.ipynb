# 02 Explore GTFS Data

## Purpose

This notebook explores METRO Houston's GTFS transit feed to understand the structure of the transit network. GTFS data provides route names, route IDs, stops, trips, schedules, and route geometry.

This step matters because route IDs from GTFS are later joined with ridership data from METRO reports. Establishing that connection allows the project to combine ridership demand with actual route geography.

## Inputs

- `data/raw/metro_gtfs/merged/routes.txt`
- `data/raw/metro_gtfs/merged/stops.txt`
- `data/raw/metro_gtfs/merged/shapes.txt`

## Outputs

- Initial understanding of METRO route IDs, stop locations, and route shape data.
- Confirmation that major routes such as Westheimer, Richmond, Bellaire, Beechnut, and Gessner are available in the GTFS feed.

## Why this notebook matters

This project depends on connecting multiple datasets that use route identifiers differently. METRO's ridership reports list routes by public-facing route numbers and names, while GTFS contains the route IDs and geography needed for mapping. This notebook confirms that those IDs can be used reliably in later analysis.


## 1. Import libraries

Pandas is used throughout this project for reading GTFS text files and inspecting the structure of the transit data.


```python
import pandas as pd
```

## 2. Load GTFS routes

The `routes.txt` file is the GTFS table that lists each transit route. This includes route IDs, short route numbers, route names, and route type.

For this project, `route_id` is especially important because it becomes the key used to join route geography, trip frequency, and ridership data.


```python
routes = pd.read_csv(
    "../data/raw/metro_gtfs/merged/routes.txt"
)

print("Routes shape:", routes.shape)
routes.head()
```

## 3. Inspect route columns

Before cleaning or joining data, it is important to understand the available columns. This helps confirm which fields should be used for route matching and later analysis.


```python
routes.columns.tolist()
```

## 4. Preview key route fields

This view narrows the route table to the fields most relevant for analysis:

- `route_id`: GTFS route identifier
- `route_short_name`: public route number
- `route_long_name`: route or corridor name
- `route_type`: mode type, where local bus routes generally use type `3`


```python
routes[
    [
        "route_id",
        "route_short_name",
        "route_long_name",
        "route_type"
    ]
].head(30)
```

## 5. Identifying key corridors

The project focuses on several high-ridership corridors that appear in METRO ridership reports. This section verifies that those corridors exist in the GTFS feed and identifies their route IDs.

The initial focus routes are:

- 82 Westheimer
- 25 Richmond
- 2 Bellaire
- 4 Beechnut
- 46 Gessner

These routes are used later for ridership, productivity, and population-density overlay analysis.


```python
key_corridors = routes[
    routes["route_long_name"]
    .astype(str)
    .str.contains("WESTHEIMER|BELLAIRE|RICHMOND|BEECHNUT|GESSNER", case=False, na=False)
]

key_corridors[
    [
        "route_id",
        "route_short_name",
        "route_long_name",
        "route_type"
    ]
].sort_values("route_id")
```

## 6. Load GTFS stops

The `stops.txt` file contains the physical stop locations in the transit network. Each stop includes latitude and longitude coordinates, which can support later accessibility and stop-density analysis.

Although this notebook does not yet calculate accessibility, confirming that stop locations are available is important for future project phases.


```python
stops = pd.read_csv(
    "../data/raw/metro_gtfs/merged/stops.txt"
)

print("Stops shape:", stops.shape)
stops.head()
```

## 7. Load GTFS shapes

The `shapes.txt` file contains the geometry points used to draw routes on a map. Each `shape_id` is made of ordered latitude/longitude points.

This table is essential for mapping routes and overlaying them with Census population-density data in later notebooks.


```python
shapes = pd.read_csv(
    "../data/raw/metro_gtfs/merged/shapes.txt"
)

print("Shapes shape:", shapes.shape)
shapes.head()
```

## 8. Summary

The GTFS feed contains usable route, stop, and shape data for METRO Houston. The key route IDs for Westheimer, Richmond, Bellaire, Beechnut, and Gessner were successfully identified, allowing future notebooks to join GTFS geography with ridership and investment metrics.

### Key takeaways

- GTFS provides the route IDs needed to connect transit geography with ridership data.
- Stop latitude and longitude data is available for future accessibility analysis.
- Shape geometry is available for mapping selected corridors.
- The identified high-ridership routes can be used in later investment-priority analysis.
