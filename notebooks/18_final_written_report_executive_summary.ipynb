Rather than focusing only on ridership totals, the project examined how efficiently routes generate riders, how well transit service aligns with population density, and where accessibility gaps persist across the region. # 3. Methodology Overview

The project was developed across a series of notebooks, each contributing a different part of the final analysis.

## GTFS Data

Houston METRO GTFS data provided route, stop, trip, and shape information.

GTFS was used to:

- Explore the transit network
- Identify routes and route types
- Build route geometries
- Count scheduled trips
- Connect routes to surrounding census tracts

This created the spatial foundation needed to analyze corridors instead of only individual route IDs.

## Census and ACS Data

American Community Survey data and Census TIGER tract boundaries were used to measure population density across Harris County.

These datasets made it possible to evaluate whether transit service aligns with where people live and where density is highest.

## METRO Ridership Reports

Twelve monthly METRO ridership reports covering October 2024 through September 2025 were processed and standardized.

The reports provided route-level productivity metrics, including:

- Average weekday boardings
- Boardings per revenue hour
- Boardings per revenue mile

These values were used to create a route efficiency summary and composite productivity score.

## Accessibility Index

A custom Transit Accessibility Index v2 was developed to measure tract-level transit access.

The index incorporated:

- Nearby stop count
- Route frequency
- Access to high-productivity routes
- Distance to METRORail

This allowed the project to identify census tracts with weaker transit accessibility relative to density and service context.

## ROI / Investment Framework

The final investment model combined:

- Route demand
- Service productivity
- Accessibility need
- Equity context
- Distance from rail
- Population exposure

These factors were integrated into a corridor-level investment priority score. The model was designed as a planning-screening framework rather than a formal engineering cost-benefit analysis.


# 4. Key Findings

## Finding 1: High ridership does not always mean high productivity

Routes with the highest ridership are not always the most efficient. Boardings per revenue hour and boardings per revenue mile provide important context for understanding how effectively service is being used.

This matters because a route may carry many riders simply because it receives a large amount of service. Another route may carry fewer total riders but generate more riders per service hour, making it highly productive.

## Finding 2: Several major bus corridors consistently perform well

Corridors such as Westheimer, Bellaire, Gessner, Fondren, Richmond, Beechnut, Bissonnet, and Hillcroft repeatedly appear as important corridors across multiple metrics.

These corridors are notable because they combine some mix of:

- High ridership
- Strong route productivity
- Large population exposure
- Accessibility need
- Distance from existing rail

## Finding 3: Accessibility gaps remain in high-density areas

The Accessibility Index v2 suggests that some high-density areas continue to experience weaker transit accessibility.

This finding is important because transit investment should not only follow existing ridership. It should also consider where unmet need exists and where better service could improve access for dense or underserved areas.

## Finding 4: Future investment decisions should consider more than ridership

Ridership alone does not capture accessibility need, equity concerns, or long-term investment potential.

A multi-factor framework produces a more complete view of corridor performance and helps distinguish between corridors that may need improved bus service, BRT investment, or eventual rail study.

# 6. Recommended Investment Types

The project recommends a tiered investment approach.

## Enhanced Bus Service

Enhanced bus service is most appropriate for corridors with moderate investment scores or corridors where demand exists but does not yet clearly justify major capital investment.

Possible improvements include:

- Increased frequency
- Improved reliability
- Stop amenities
- Signal priority
- Better transfers
- Longer service spans

## Bus Rapid Transit

BRT may be appropriate for corridors with strong demand, productivity, population exposure, and accessibility need.

BRT can provide many benefits associated with rail-like service while generally requiring lower capital investment than light rail.

Potential BRT candidates include:

- Richmond
- Beechnut
- Bissonnet
- Hillcroft
- Gulfton-area corridors

## Rail / High-Capacity Transit Evaluation

Rail should be considered only for the strongest long-term corridors where multiple indicators align.

Potential rail or high-capacity transit study candidates include:

- Westheimer
- Bellaire
- Gessner
- Fondren

These corridors require additional engineering, cost, right-of-way, and political feasibility analysis before any final recommendation could be made.


# 7. Policy Implications

The analysis suggests that Houston's future transit investments should prioritize corridors where demand, accessibility need, and population exposure overlap.

A tiered investment approach is more realistic than assuming every strong corridor requires rail. Some corridors may benefit most from improved bus service, while others may warrant BRT or rail study.

The results also suggest that accessibility and equity metrics should be included alongside traditional ridership measures. Without these additional factors, investment decisions may over-prioritize already successful corridors while missing areas of unmet transit need.


# 8. Limitations

This project uses public GTFS, Census, and METRO ridership data to create a planning-style screening model.

Important limitations include:

- Route buffers approximate transit access and do not model actual pedestrian network distance.
- Population exposure may double-count overlapping census tracts.
- Employment density was not included.
- Zero-vehicle household data was not incorporated.
- Capital construction costs were not explicitly modeled.
- Travel time savings were not calculated.
- Investment recommendations represent screening-level guidance rather than engineering conclusions.
- Rail recommendations would require right-of-way, funding, political, and environmental review.

Despite these limitations, the project provides a transparent and data-driven framework for comparing corridors using available public data.

# 10. Conclusion

The final recommendation is not simply to build rail on the busiest corridors. Instead, the analysis supports a tiered investment strategy:
