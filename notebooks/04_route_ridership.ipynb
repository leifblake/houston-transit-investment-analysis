# 04 Route Ridership

## Purpose

This notebook creates an initial route-level ridership dataset for selected high-ridership METRO Houston bus corridors.

The METRO ridership reports are published as PDFs, so this notebook manually enters a small set of route-level ridership values from the October 2024 report. This early sample is used to validate that METRO ridership data can be joined successfully to GTFS route IDs.

## Inputs

- `data/raw/metro_gtfs/merged/routes.txt`
- METRO October 2024 Ridership Report

## Outputs

- `data/processed/route_ridership.csv`

## Why this matters

Ridership represents transit demand. By joining route-level ridership to GTFS route IDs, later notebooks can combine demand, service frequency, route geography, and population density into a transit investment analysis.


## 1. Import libraries and load GTFS route data

The GTFS `routes.txt` file provides the official METRO route IDs and names. These route IDs are used as the join key for combining GTFS data with ridership data.


```python
import pandas as pd

routes = pd.read_csv(
    "../data/raw/metro_gtfs/merged/routes.txt"
)

routes.head()
```

## 2. Enter route-level ridership values

This sample uses five major bus routes from the October 2024 METRO ridership report. The values represent average weekday boardings.

These routes were selected because they are major corridors and appeared as strong candidates during early project exploration:

- Route 82 — Westheimer
- Route 46 — Gessner
- Route 25 — Richmond
- Route 2 — Bellaire
- Route 4 — Beechnut


```python
ridership = pd.DataFrame({
    "route_id": [
        82,
        46,
        25,
        2,
        4
    ],
    "avg_weekday_boardings": [
        14054,
        7670,
        7489,
        7590,
        8743
    ]
})

ridership
```

## 3. Prepare route IDs for merging

Route IDs need to have the same data type in both datasets. This step converts the GTFS route ID column into numeric values so it can be joined with the manually entered ridership table.


```python
routes["route_id"] = pd.to_numeric(
    routes["route_id"],
    errors="coerce"
)

merged = routes.merge(
    ridership,
    on="route_id",
    how="inner"
)

merged
```

## 4. Visualize average weekday boardings

This chart provides a quick comparison of weekday ridership across the selected corridors. It helps identify which routes carry the most riders before service frequency or population density are considered.


```python
import matplotlib.pyplot as plt

merged.sort_values(
    "avg_weekday_boardings",
    ascending=False
).plot(
    x="route_long_name",
    y="avg_weekday_boardings",
    kind="bar",
    figsize=(10, 5)
)

plt.title("Average Weekday Boardings by Selected METRO Route")
plt.ylabel("Average Weekday Boardings")
plt.xlabel("Route")
plt.tight_layout()
plt.show()
```

## 5. Save the processed ridership dataset

The cleaned route ridership file is saved to `data/processed/` so later notebooks can reuse it without repeating the manual entry step.


```python
merged.to_csv(
    "../data/processed/route_ridership.csv",
    index=False
)

print("Saved route_ridership.csv")
```

## Summary

This notebook created the first processed ridership dataset for the project. The selected routes were successfully joined to GTFS route metadata, confirming that METRO ridership values can be connected to GTFS route IDs.

The resulting dataset supports later analysis of route productivity, service frequency, geographic coverage, and investment priority.
